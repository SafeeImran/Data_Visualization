```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import Normalize
from adjustText import adjust_text


# Calculate label coordinates
shx['label_coordinates'] = shx['geometry'].centroid

# Define the colormap
cmap = cm.get_cmap('magma')

# Normalize the values
norm = Normalize(vmin=df4_geo['Population '].min(), vmax=df4_geo['Population '].max())

# Plot the map and city labels
fig, ax = plt.subplots(figsize=(10, 10))
shx.plot(ax=ax, color='lightblue', edgecolor='black')

labels = []
for x, y, label, value in zip(shx['label_coordinates'].x, shx['label_coordinates'].y, df4_geo['Cities'], df4_geo['Population ']):
    label_text = ax.text(x, y, label, fontsize=16, ha='center', color=cmap(norm(value)), weight = "bold")
    labels.append(label_text)

# Adjust label positions to avoid overlap
adjust_text(labels, ax=ax)

# Create a colorbar for the colormap
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, ax=ax)

plt.show()
```
